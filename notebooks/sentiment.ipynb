```python
def parser(link_path):
  with open(link_path, 'r') as file:
      data = json.load(file)

  links = [entry['link'] for entry in data]

  unique_links = list(set(links))

  return unique_links
```

```python
def fetch_article_content(link):
    article = Article(link)
    article.download()
    article.parse()
    return article
```

```python
def generate_newspaper(article_links):
    newspaper_content = ""
    for link in article_links:
        article = fetch_article_content(link)
        newspaper_content += f"Title: {article.title}\n"
        newspaper_content += f"Authors: {', '.join(article.authors)}\n" if article.authors else ""
        newspaper_content += f"Publication Date: {article.publish_date}\n" if article.publish_date else ""
        newspaper_content += f"Text: {article.text}\n\n"
    return newspaper_content
```

```python
newspaper_content = generate_newspaper(article_links)
print(newspaper_content)
```

```python
def extract_text(link):
    article = fetch_article_content(link)

    return article.text.strip()
```

```python
raw_text = extract_text("https://www.tennessean.com/story/news/crime/2023/12/20/two-charged-year-long-probe-nashville-toddler-ariel-rose-death/71991923007")
```

```python
import re

def split_and_clean_lines(input_string):
    lines = input_string.split('\n')
    non_empty_lines = [re.sub(r'[^\w\s\'\.\-]', '', line.strip()) for line in lines if line.strip()]
    return non_empty_lines
```

```python
clean_text = split_and_clean_lines(raw_text)
print(clean_text)
```

```python
def write_text(text, path):
    with open(path, 'a') as file:
        for line in text:
            file.write(line + '\n')
```

```python
from newspaper import ArticleException


def process_location_links(links, text_path):
    for link in links:
        try:
            raw_text = extract_text(link)
            clean_text = split_and_clean_lines(raw_text)
            write_text(clean_text, text_path)
        except ArticleException:
            print("Skipping " + link)
```

```python
import pandas as pd
```

```python
with open(nashville_text_path, 'r') as file:
    lines = file.readlines()

nashville_entries = pd.DataFrame(lines, columns=['content'])

print(nashville_entries)
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer as SIA

sia = SIA()
sia.polarity_scores("This restaurant was great, but I'm not sure if I'll go there again.")
```

```python
def get_scores(content):
    sia_scores = sia.polarity_scores(content)

    return pd.Series({
        'nltk': sia_scores['compound'],
    })
```

```python
scores = nashville_entries.content.apply(get_scores)
```

```python
print(scores)
```

```python
import matplotlib.pyplot as plt

plt.hist(scores['nltk'], bins=20)  # Adjust the number of bins as needed
plt.title('Distribution of ' + 'Nashville Sentiment')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
def score_entries(text_path):
    with open(text_path, 'r') as file:
        lines = file.readlines()
    
    entries = pd.DataFrame(lines, columns=['content'])
    
    return entries.content.apply(get_scores)
```

```python
def plot_scores(entries, city):
    plt.hist(entries['nltk'], bins=20)  # Adjust the number of bins as needed
    plt.title(f'Distribution of ' + city + ' Sentiment')
    plt.xlabel('Values')
    plt.ylabel('Frequency')
    plt.show()
```

```python
nashville_scores = score_entries(nashville_text_path)
plot_scores(nashville_scores, "Nashville")
```

```python
memphis_scores = score_entries(memphis_text_path)
plot_scores(memphis_scores, "Memphis")
```

```python
knoxville_scores = score_entries(knoxville_text_path)
plot_scores(knoxville_scores, "Knoxville")
```

```python
def filter_scores(scores):
    return scores[scores['nltk'] != 0]
```

```python
nashville_filtered_scores = filter_scores(nashville_scores)
plot_scores(nashville_filtered_scores, "Nashville Filtered")
```

```python
memphis_filtered_scores = filter_scores(memphis_scores)
plot_scores(memphis_filtered_scores, "Memphis Filtered")
```

```python
knoxville_filtered_scores = filter_scores(knoxville_scores)
plot_scores(knoxville_filtered_scores, "Knoxville Filtered")
```

```python
df = pd.DataFrame(nashville_filtered_scores['nltk'])

# Calculate frequencies
value_counts = df['nltk'].value_counts()

# Calculate total frequency
total_frequency = value_counts.sum()

# Calculate percentages
percentages = (value_counts / total_frequency) * 100

# Create the plot
plt.bar(percentages.index, percentages.values)
plt.xlabel('Values')
plt.ylabel('Percentage')
plt.title('Distribution of Data with Respect to the Whole Set')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
def percentage_distribution(filtered_scores, place):
    df = pd.DataFrame(filtered_scores['nltk'])

    value_counts = df['nltk'].value_counts()
    
    total_frequency = value_counts.sum()
    
    percentages = (value_counts / total_frequency) * 100
    
    plt.bar(percentages.index, percentages.values)
    plt.xlabel('Values')
    plt.ylabel('Percentage')
    plt.title('Percentage Distribution of Filtered ' +  place + ' Scores')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

```python
percentage_distribution(nashville_filtered_scores, "Nashville")
```

```python
percentage_distribution(memphis_filtered_scores, "Memphis")
```

```python
percentage_distribution(knoxville_filtered_scores, "Knoxville")
```
